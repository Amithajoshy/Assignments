# Feature engineering

```python
data['avg_trails_watched_per_day']=data['slot1_trails_watched_per_day']+data['slot2_trails_watched_per_day']+data['slot3_trails_watched_per_day']+data['slot4_trails_watched_per_day']
data['avg_trails_watched_per_day']=data['avg_trails_watched_per_day']/4
data
```

```python
data=data.drop(columns="slot1_trails_watched_per_day")
data=data.drop(columns="slot2_trails_watched_per_day")
data=data.drop(columns="slot3_trails_watched_per_day")
data=data.drop(columns="slot4_trails_watched_per_day")
data
```

```python
data.corr()
```

```python
from sklearn.preprocessing import StandardScaler
variables=['Unnamed: 0', 'userId', 'tier', 'gender', 'following_rate',
       'followers_avg_age', 'following_avg_age', 'max_repetitive_punc',
       'num_of_hashtags_per_action', 'emoji_count_per_action',
       'punctuations_per_action', 'number_of_words_per_action',
       'avgCompletion', 'avgTimeSpent', 'avgDuration', 'avgComments',
       'creations', 'content_views', 'num_of_comments',
       'weekends_trails_watched_per_day', 'weekdays_trails_watched_per_day',
       'avgt2','avg_trails_watched_per_day']
x=data.loc[:,variables].values
y=data.loc[:,['age_group']].values
x=StandardScaler().fit_transform(x)
x=pd.DataFrame(x)
x
```

```python
data.columns
```

```python

from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
ridge=Ridge(alpha=0.1,normalize=True)
ridge.fit(x_train,y_train)
print('Ridge Regression')
print('R^2 score for train:',ridge.score(x_train, y_train))
print('R^2 score for test:',ridge.score(x_test, y_test))
y_pred = ridge.predict(x_test)
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
#Accuracy 
Accuracy=r2_score(y_test,y_pred)*100
print('Accuracy: ',Accuracy) 
```

```python
ridge=Ridge(alpha=0.0000000001,normalize=True)
ridge.fit(x_train,y_train)
print('Ridge Regression-2')
print('R^2 score for train:',ridge.score(x_train, y_train))
print('R^2 score for test:',ridge.score(x_test, y_test))
y_pred = ridge.predict(x_test)
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
#Accuracy 
Accuracy=r2_score(y_test,y_pred)*100
print('Accuracy: ',Accuracy) 
```
