# visualisation

```python
plt.figure(figsize=(8,8))
plt.hist(data['SL'],color="red",rwidth=0.9,density=True)
plt.title("Distribution of SL",fontsize=16)
plt.xlabel('SL')
plt.ylabel('frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(data['PL'],data['PW'],s=12,c='red')
plt.title('plot of PL VS PW',fontsize=16)
plt.xlabel('PL')
plt.ylabel('PW')
```

```python
sns.stripplot(data['SW'])
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['SW'],color="red",rwidth=0.9,density=True)
plt.title("Distribution of SW",fontsize=16)
plt.xlabel('SW')
plt.ylabel('frequency')
plt.xticks()
plt.yticks()
```
